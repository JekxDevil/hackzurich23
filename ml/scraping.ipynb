```python
import os

to_parse = []
for f in os.listdir('data'):
    if f.endswith('ps1'):
        to_parse.append(f)

for fpath in to_parse:
    with open('data/' + fpath, 'r', encoding="utf8", errors="replace") as f:
        print()
        print(fpath)
        s = f.read()
        # retrieve last 4 lines from s
        last_lines = s.split('\n')[-6:]
        for line in last_lines:
            colon = line.find(':')
            if colon == -1:
                continue
            print(line)
            data = line[colon + 1:].strip()
            pos = s.find(data)
            print(pos, pos + len(data))
```

```python
with open("data/community-five-go.xml", "r", encoding="utf8", errors="replace") as f:
    s = f.read()
    lines = s.split('\n')
    prev = -1
    for line in lines:
        if "name" in line or "address" in line or "iban" in line:
            parts = line.split('>')
            data = parts[1].split('<')[0].strip()
            if len(data) == 0:
                continue
            typ = "person" if "name" in line else "address" if "address" in line else "iban"
            pos = s.find(data)
            if (pos < prev):
                continue
            prev = pos
            print(f"({pos}, {pos + len(data)}, \"{typ.upper()}\"),")
```
